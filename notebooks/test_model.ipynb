```python
import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np
from typing import List
from phd_visualizations.calculations import calculate_uncertainty
from typing import Literal

supported_instruments = Literal['pt100', 'pt1000', 'humidity_capacitive', 'vortex_flow_meter', 'paddle_wheel_flow_meter']

def regression_plot(df_ref: pd.DataFrame, df_mod: pd.DataFrame, var_ids: List[str], instruments: List[supported_instruments] = None):
    if instruments is None:
        instruments = [None] * len(var_ids)
    
    fig = make_subplots(rows=len(var_ids), cols=1, subplot_titles=[f'Regression plot for {var_id}' for var_id in var_ids])
    
    for i, var_id in enumerate(var_ids):
        x = df_ref[var_id].values.reshape(-1, 1)
        y = df_mod[var_id].values
        
        # Perform linear regression
        model = LinearRegression()
        model.fit(x, y)
        y_pred = model.predict(x)
        
        # Calculate RMSE
        rmse = np.sqrt(mean_squared_error(y, y_pred))
        
        # Calculate uncertainty
        if instruments[i]:
            uncertainty = calculate_uncertainty(y_pred, instruments[i])
        else:
            uncertainty = np.zeros_like(y_pred)
        
        # Create scatter plot
        scatter = go.Scatter(x=df_ref[var_id], y=df_mod[var_id], mode='markers', name='Data')
        
        # Create regression line plot
        regression_line = go.Scatter(x=df_ref[var_id], y=y_pred, mode='lines', name='Regression Line')
        
        # Create uncertainty plot
        upper_bound = go.Scatter(x=df_ref[var_id], y=y_pred + uncertainty, mode='lines', fill=None, line=dict(color='lightgrey'), showlegend=False)
        lower_bound = go.Scatter(x=df_ref[var_id], y=y_pred - uncertainty, mode='lines', fill='tonexty', line=dict(color='lightgrey'), showlegend=False)
        
        # Add plots to subplot
        fig.add_trace(scatter, row=i+1, col=1)
        fig.add_trace(regression_line, row=i+1, col=1)
        fig.add_trace(upper_bound, row=i+1, col=1)
        fig.add_trace(lower_bound, row=i+1, col=1)
        
        # Update subplot title with RMSE
        fig.update_yaxes(title_text=f'RMSE: {rmse:.2f}', row=i+1, col=1)
    
    fig.update_layout(height=300*len(var_ids), title_text="Regression Plots with Uncertainty")
    
    return fig
```
