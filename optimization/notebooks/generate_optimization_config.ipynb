# Configuration generation

Notebook to generate file of simulations configuration for [experimental_horizon_optimization DAG](../../deployment/dags/experimental_horizon_optimization.py) 

```python
from pathlib import Path
from solhycool_optimization import (EvaluationConfig, 
                                    ModelInputsRange, 
                                    MatlabOptions, 
                                    ValuesDecisionVariables,
                                    AlgoParamsHorizon as AlgoParams)

output_path = Path("/workspaces/SOLhycool/optimization/data/optimization_config.json")
```

### Pilot plant. Combined cooler

```python
# Pilot plant. CC
model_inputs_range = ModelInputsRange(
    qc = (6., 24.),
    Rp = (0.0, 1.0),
    Rs = (0.0, 1.0),
    wdc = (11, 99.18),
    wwct = (21, 93.4161),
    Tamb = (0.1, 50),
    HR = (0.1, 99.9),
    Tdc_in = (25, 65),
    Twct_in = (25, 65)
)

config = EvaluationConfig(
    id="pilot_plant",
    description="For experimental validation of combined cooler in pilot plant at PSA",
    model_inputs_range=model_inputs_range,
    matlab_options=MatlabOptions.from_model_inputs_range(
        model_inputs_range,
        other_options={
            "wct_model_data_path": "/workspaces/SOLhycool/modeling/data/models_data/model_data_wct_fp_pilot_plant_gaussian_cascade.mat",
            "wct_n_wct": 1,
            "dc_model_data_path": "/workspaces/SOLhycool/modeling/data/models_data/model_data_dc_fp_pilot_plant_gaussian_cascade.mat",
            "dc_n_dc": 1,
            "model_type": "data",
            "silence_warnings": True,
        },
    ),
    algo_params=AlgoParams(),
    vals_dec_vars=ValuesDecisionVariables(
        qc=7,
        Rp=6,
        Rs=6,
        wdc=10,
    ),
)

config.to_config_file(output_path)
```
