```python
from pathlib import Path
import hjson
import pandas as pd
from loguru import logger

from phd_visualizations.test_timeseries import experimental_results_plot
from phd_visualizations import save_figure
from solhycool_visualization.analysis import year_pie_plot

# auto reload modules
%load_ext autoreload
%autoreload 2

data_path: Path = Path("../../data")
results_path: Path = Path("/workspaces/SOLhycool/simulation/results/")

# 20220201_20220228/cc_static/df_sim_eval_at_20250401T0814_andasol.h5
file_path: Path = results_path / "20220101_20221231/cc_static/df_sim_eval_at_20250331T1959_andasol.h5"


df = pd.read_hdf(file_path)
for var_id in ["Tv", "Tc_in", "Tc_out"]: # TODO: This should be provided directly by OperationPoint.null
    df.loc[df[var_id] < 5, var_id] = None
df
```

```python
# If resulting file is too big, resave data compressed
df.to_hdf(file_path, key="simulation_data", format="table", complib="zlib", complevel=9, mode="w")
logger.info(f"File saved to {file_path}")
```

```python
# Exportar datos a csv para que se puedan importar en otros entornos
df.to_csv(file_path.with_suffix(".csv"))
```

```python
# Resample results monthly
df = df.resample("SME").mean()
df
```

```python
plot_config = hjson.load(open(data_path / "plot_config_year.hjson"))
plot_config["subtitle"] = f"Evaluation results for {results_path.parts[-2].replace('_', ' ')}"

# Move contents of load conditions to cooling power and remove it
plot_config["plots"]["cooling_power_distribution"]["ylims_left"] = [0,250]
plot_config["plots"]["cooling_power_distribution"]["traces_right"] = plot_config["plots"]["load_conditions"]["traces_right"]
plot_config["plots"]["cooling_power_distribution"].update({param_id: plot_config["plots"]["load_conditions"][param_id] for param_id in ["ylabels_right", "ylims_right"]})
plot_config["plots"].pop("load_conditions")

fig = experimental_results_plot(
    plot_config, 
    df=df, # best_dfs[0],  
    resample=False,
)
# fig.update_layout(width=1000, height=1000)

fig
```

```python
save_figure(
    fig,
    figure_name=file_path.stem,
    # figure_name=f"{df.index[0]:%Y%m%d}_{df.index[1]:%Y%m%d}_optim_viz",
    # figure_path=Path("../results"),
    figure_path=file_path.parent,
    formats=["png", "html"],
)
```
